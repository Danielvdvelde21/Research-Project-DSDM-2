## Atopic Asthma

*"Type of asthma, possible that children with atopic asthma react more to pollen."*<br>
Just putting the ones with atopic asthma and the ones without it on scatter plots, doesn't show correlation.

```python
asthma['AtopicAsthma'].value_counts()
```

```python
kids_with_atopic = asthma.loc[asthma['AtopicAsthma'] == 'Yes', ['pollenTotal', 'ACD6score']]
kids_with_atopic.plot(kind='scatter', x='pollenTotal', y='ACD6score', figsize=(6,6))
```

```python
kids_without_atopic = asthma.loc[asthma['AtopicAsthma'] == 'No', ['pollenTotal', 'ACD6score']]
kids_without_atopic.plot(kind='scatter', x='pollenTotal', y='ACD6score', figsize=(6,6))
```
